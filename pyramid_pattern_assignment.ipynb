```python
n=int(input("Enter the value of n"))
x=1
for i in range(n):
    for j in range(n-1-i):
        print(" ",end=" ")
    if(i%2==0):
        for k in range(2*i+1):
            print(x,end=" ")
        x+=1
    else:
        for k in range(2*i+1):
            print("*",end=" ")
    print()
```

```python
n=int(input("Enter the value of n"))
x=1
p=True
for i in range(n):
    for j in range(n-1-i):
        print(" ",end=" ")
    for k in range(2*i+1):
        if(p==True):
            print(x,end=" ")
            x+=1
            p=False
        else:
            print("*",end=" ")
            p=True
    print()
```
