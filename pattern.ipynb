```python
row=int(input("Enter the number of rows"))
col=int(input("Enter the number of columns"))                             # holds the ascii value of 'A'
for i in range (row):
    val=ord('A')
    for j in range (col) :
        print(chr(val), end=" ") 
        val+=1        # chr(ascii_value i.e. 'val' ) converts to character
    print()
```

```python
row=int(input("Enter the number of rows"))
col=int(input("Enter the number of columns"))
for i in range (row):
    val=ord('A')
    for j in range (col):
        print(chr(val) , end=" ")
        val+=1
    print()
    if val>ord('Z'):
        val=ord('A')
```

```python
n=int(input("Enter a number"))
val=1
p=True
for i in range(n):
    for j in range(n):
        if p:
            print(val ,end=" ")
            val+=1
            if val>9:val=1
            p=False
        else:
            print("*",end=" ")
            p=True
    print()
```

```python
n=int(input("Enter a number"))
val=1
# p=True
for i in range(n):
    for j in range(n):
        if i%2==0:
            print(val ,end=" ")
            if val>9:val=1
            # p=False
        else:
            print("*",end=" ")
            # p=True
    print()
    if i%2==0:
        val+=1
```
