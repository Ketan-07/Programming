# Linear Search

## Time Complexity :-
* Best Case - O(1)
* Worst Case - O(N)

```python
li=[10,2,5,8,3,4]
key=int(input("n:"))
bool=False
for i in range(len(li)):
    if key==li[i]:
        print(f"{key} is found at the index {i} and position {i+1}")
        bool=True       
        break
if not bool:
    print("Key not found in the list")
```

# Linear Search using - For-Else
* we can use else and for under same indentation when for loop dosen't executes full and break in middle and we need to print some extra messages or do some activity

```python
li=[10,2,5,8,3,4]
key=int(input("n:"))
for i in range(len(li)):
    if key==li[i]:
        print(f"{key} is found at the index {i} and position {i+1}")    
        break
else:
    print("Key not found in the list")
```

# Linear Search for Second Occurence

```python
li=[10,2,5,8,3,4,6,7,3,4,2,1,4,5,6,9]
key=int(input("n:"))
bool=False
count=0
for i in range(len(li)):
    if key==li[i]:
        count+=1
        if count==2:
            print(f"{key} is found at the index {i} and position {i+1}")
            bool=True       
            break
if not bool:
    if count>0:
        print(f"{key} found {count} times but not full")
    else:
        print("Key not found in the list")
```

# Linear Search for Nth Occurence

# Binary Search
* Divide and Conquer Method

```python
li=[10,2,5,8,3,4]
key=int(input("key:"))
first=0
last=len(li)-1
bool=False
while first<=last:
    mid=(first+last)//2
    if key<li[mid]:
        last=mid-1
    elif key>li[mid]:
        first=mid+1
    else:
        print(f'{key} is found at index {mid}')
        bool=True
        break

if not bool:
    print(f"{key} is not found")
```
