# Write a program to count vowels and consonants

```python
s1='eDucAtion iS PoweR'
vowel=0
consonant=0
for i in s1:
    if i in "AEIOUaeiou":
        vowel+=1
    elif (65<=ord(i)<=90 or 97<=ord(i)<=122) and i not in "AEIOUaeiou":
        consonant+=1
print('Vowels - ',vowel)
print('consonant - ',consonant)
```

# Converting Lower to upper and Upper to Lower

```python
s1='eDucAtion iS PoweR'
s2=''
print(s1)
for i in s1:
    if i==0:
        s2+=1
    elif ord(i) >= 65 and ord(i) <= 90:
        s2+=chr(ord(i)+32)
    elif ord(i) >=97 and ord(i) <= 122:
        s2+=chr(ord(i)-32)

print(s2)
```

# Write a program to convert Camel case to Snake Case
* Input - "If You Are Good At Something Never Do It For Free"
* Output - "if_you_are_good_at_something_never_do_it_for_free"

```python
s1="If You Are Good At Something Never Do It For Free"
s2=''
print(s1)
for i in s1:
    if i==0:
        s2+=1
    elif ord(i) >= 65 and ord(i) <= 90:
        s2+=chr(ord(i)+32)
    elif ord(i) == 32:
        s2+='_'
    else:
        s2+=i
print(s2)
```

# Write a program to convert Snake case to Camel Case
* Input - "if_you_are_good_at_something_never_do_it_for_free"
* Output - "If You Are Good At Something Never Do It For Free"

```python
s1="if_you_are_good_at_something_never_do_it_for_free"
s2=''
s2 += chr(ord(s1[0])-32)
print(s1)

for i in range(1,len(s1)):
    if s1[i] == '_':
        s2+=" "
    else:
        if s1[i-1] == "_":
            s2 += chr(ord(s1[i])-32)
        else:
            s2 += s1[i]
print(s2)
```

# Write a program to check whether a string is palindrome or not

```python
# Check the given string is palindrom of not
s1='madam'
s2=''
for i in range(len(s1)-1,-1,-1):
    s2+=s1[i]
if s1==s2:
    print('Palindrome')
else:
    print('Not Palindrome')
```

# Accept two strings and mix it as follows:
* s1="abcd"
* s2="wxyz"

* Output = "azbycxdw"

```python
s1="abcd"
s2="wxyz"
s3=''
j=len(s1)-1
for i in range(len(s1)):
    s3+=(s1[i]+s2[j])
    j-=1
print(s3)
```

```python
s1="abcd"
s2="wxyz"
s3=''
for i in range(len(s1)):
    s3+=(s1[i]+s2[len(s2)-i-1])
print(s3)
```

# Write a program for given pattern 
* Input - 'pyspiders'
* Output - if even length then no output else output in plus format

```python
s1 = 'pyspiders'
mid=len(s1)//2
for i in range(len(s1)):
    for j in range(len(s1)):
        if j == mid:
            print(s1[i],end=' ')
        elif i==mid:
            print(s1[j],end=' ')
        else:
            print(' ',end=' ')
    print()
```
