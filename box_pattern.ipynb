```python
n=int(input("Enter a number"))
mid=n//2
for i in range(n):
    for j in range(n):
        if(i==0 or j==0 or i==(n-1) or j==(n-1) or (i==mid and j==mid)):
            print("*",end=" ")
        else:
            print(" ",end=" ")
    print()
```
