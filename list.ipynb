# Find the sum of prime number given in a list using functions
* input - [2,1,3,6,5,1,4]
* output - 10(2+3+5)

```python
def isPrime(n):
    c=0
    for i in range(2,(n//2)+1):
        if(n%i==0):
            c+=1
    if c==0 and n!=1:
        return True
    else:
        return False

li=[2,1,3,6,5,1,4]
sum=0
for i in li:
    if isPrime(i):
        sum+=i
print(sum)
```

```python
def isPrime(n):
    if n==1:
        return False
    for i in range(2,n):
        if n%i==0:
            return False
    return True

sum=0
li=[2,1,3,6,5,1,4]
for i in range(len(li)):
    if isPrime(li[i]):
        sum+=li[i]
print(f'Sum of prime elements from the list is :',sum)
```
