# Removing a Particular Element from the list

```python
li=[1,2,3,1,1,1,1,4,5]
target=int(input("Targer:"))
for i in li:
    if i==target:
        li.remove(i)
print(li,"\n")
print('''The above code is not working as expected because when we remove an element from the list,
the index of the next element gets shifted to the left. 
So, if there are multiple occurrences of the target element in the list, 
it will only remove the first occurrence and skip the rest.
To fix this issue, we can use a while loop instead of a for loop to iterate through the list and remove all occurrences of the target element.
''')
```

```python
li=[1,2,3,1,1,1,1,4,5]
print("Original List",li,'\n')
print("List After Removing Duplicates")
for i in range(len(li)):
    for j in range((i+1),len(li)):
        if(li[j]==li[i]):
            li[j]=''
while '' in li:
    li.remove('')
print(li)
```

# List rotation

### Clockwise or Leftside Rotation

```python
li = [1,2,3,4,5,6]
temp=li[0]
for i in range(len(li)-1):
    if i==len(li)-1:
        li[i]=temp
    else:
        li[i]=li[i+1]
print(li)
```

* Single Rotation

```python
n=int(input('n:'))
li=list(map(int,input().split()))[:n]
temp=li[0]
for i in range(len(li)-1):
    li[i]=li[i+1]
li[-1]=temp
print(li)
```

* K Rotation

```python
n=int(input('n:'))
li=list(map(int,input().split()))[:n]
k=int(input("Enter K for left rotation"))
loop_counter=0
for val in range(k):
    loop_counter+=1
    temp=li[0]
    for i in range(len(li)-1):
        li[i]=li[i+1]
    li[-1]=temp
print(li,loop_counter)
```

* Reduced time complexity And loop_Counter

```python
n=int(input('n:'))
li=list(map(int,input().split()))[:n]
k=int(input("Enter K for left rotation"))
loop_counter=0
for val in range(k%len(li)):
    loop_counter+=1
    temp=li[0]
    for i in range(len(li)-1):
        li[i]=li[i+1]
    li[-1]=temp
print(li,loop_counter)
```

* Rotation Using Slicing

```python
n=int(input('n:'))
li=list(map(int,input().split()))[:n]
k=int(input("Enter K for left rotation"))
for val in range(k%len(li)):
    li=li[1:]+li[:1]
print(li)
```

### Anticlock Wise or Right side Rotation

* Single rightside Rotation

```python
n=int(input('n:'))
li=list(map(int,input().split()))[:n]
temp=li[-1]
for i in range(len(li)-1,0,-1):
    li[i]=li[i-1]
li[0]=temp
print(li)
```

* For loop Rightside K Rotation

```python
n=int(input('n:'))
li=list(map(int,input().split()))[:n]
k=int(input("Enter K for left rotation"))
loop_counter=0
for val in range(k):
    loop_counter+=1
    temp=li[-1]
    for i in range(len(li)-1,0,-1):
        li[i]=li[i-1]
    li[0]=temp
print(li,loop_counter)
```
