# Even Odd Series

```python
n=int(input("Enter number of terms required"))
even=2
odd=1
for i in range(n):
    if i%2==0:
        print(even,end=" ")
        even+=2
    else:
        print(odd,end=" ")
        odd+=2
```

```python
n=int(input("Enter number of terms required"))
for i in range(n):
    if i%2==0:
        print(i+2,end=" ")
    else:
        print(i,end=" ")
```

**Ternary Operator**

```python
n=int(input("Enter number of terms required"))
for i in range(n):
    print(i+2 if i%2==0 else i ,end=" ")
```

```python
n=int(input("Enter number of terms required"))
val=1
for i in range(n):
    if i%2==0:
        print(val,end=" ")
        val+=4
    else:
        print(val,end=" ")
        val-=2
```

```python
n=int(input("Enter number of terms required"))
for i in range(n):
    print(i+1 if i%2==0 else i+4 ,end=" ")
```

# Fibonacci Series

```python
n=int(input("Enter number of terms required"))
a=0
b=1
c=0
for i in range(n):
    print(a,end=" ")
    c=a+b
    a=b
    b=c
```

```python
n=int(input("Enter number of terms required"))
a,b=0,1
for i in range(n):
    print(a,end=" ")
    c=a+b
    a,b=b,c
```

```python
n=int(input("Enter number of terms required"))
a,b=0,1
for i in range(n):
    print(a,end=" ")
    c=a+b
    a,b=b,a+b
```
