# Bubble Sort / In-place Sorting Algorithm

* Placing the first largest element to its last position
* It is a stable algoritm
* Number of Comparison in worst case(Descending Number) - (n*(n-1))/2

### Time Complexity :-
* Best Case :- O(n)
* Worst Case / Avg Case :- O(n^2)

```python
li=[5,4,3,2,1]
for i in range(len(li)-1):
    for j in range(len(li)-1-i):
        if li[j]>li[j+1]:
            li[j],li[j+1]=li[j+1],li[j]
print(li)
```

# Insertion Sort

```python
li=[23,1,10,5,2]
print("Unsorted list - ",li)
for i in range(len(li)):
    key=li[i]
    j=i-1

    while j>0 and key<li[j]:
        li[j+1]=li[j]
        j=j-1
    li[j+1]=key

print("Sorted Array",li)
```
