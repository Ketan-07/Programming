# Day 1

# swap 2 values


```python
# with an extra variable
a=10
b=20
print(f'a={a}')
print(f'b={b}')

temp=a
a=b
b=temp

print(f'a={a}')
print(f'b={b}')
```

```python
# without extra variable using operator
a=int(input('a:'))
b=int(input('b:'))
print(f'Before Swap \n a:{a}\t b:{b}')
a=a+b
b=a-b
a=a-b
print(f'After Swap \n a:{a}\t b:{b}')
```

```python
# without extra variable using operator
# It will work for negative and positive but not for zero
a=int(input('a:'))
b=int(input('b:'))
print(f'Before Swap \n a:{a}\t b:{b}')
a=a*b
b=a//b
a=a//b
print(f'After Swap \n a:{a}\t b:{b}')
```

```python
# without extra variable and without operator
a=int(input('a:'))
b=int(input('b:'))
print(f'Before Swap \n a:{a}\t b:{b}')

a,b=b,a

print(f'After Swap \n a:{a}\t b:{b}')
```

```python
# Using Bitwise operator XOR
a=int(input('a:'))
b=int(input('b:'))
print(f'Before Swap \n a:{a}\t b:{b}')

a=a^b
b=a^b
a=a^b
print(f'After Swap \n a:{a}\t b:{b}')
```

# swapping of three values

```python
# with an extra variable
a=int(input('a:'))
b=int(input('b:'))
c=int(input('c:'))
print(f'Before Swap \n a:{a}\t b:{b}\t c:{c}')

temp=a
a=b
b=c
c=temp

print(f'After Swap \n a:{a}\t b:{b}\t c:{c}')
```

```python
a=int(input('a:'))
b=int(input('b:'))
c=int(input('c:'))
print(f'Before Swap \n a:{a}\t b:{b}\t c:{c}')
a=a+b+c
b=a-(b+c)
c=a-(b+c)
a=a-(b+c)
print(f'After Swap \n a:{a}\t b:{b}\t c:{c}')
```

```python
a=int(input('a:'))
b=int(input('b:'))
c=int(input('c:'))
print(f'Before Swap \n a:{a}\t b:{b}\t c:{c}')
a=a*b*c
b=a//(b*c)
c=a//(b*c)
a=a//(b*c)
print(f'After Swap \n a:{a}\t b:{b}\t c:{c}')
```

```python
a=int(input('a:'))
b=int(input('b:'))
c=int(input('c:'))
print(f'Before Swap \n a:{a}\t b:{b}\t c:{c}')
a=a^b^c
b=a^b^c
c=a^b^c
a=a^b^c
print(f'After Swap \n a:{a}\t b:{b}\t c:{c}')
```

# Even Odd Number

```python
n=int(input('Enter a Number'))
if(n%2==0):
    print(f"{n} is Even Number")
else:
    print(f"{n} is Odd Number")
```

```python
# using temp variable
n=int(input('Enter a Number'))
temp=n
temp=temp//2
temp=temp*2
if(temp==n):
    print(f'{n} is Even Number')
else:
    print(f"{n} is Odd Number")
```

```python
# without using extra temp variable
n=int(input('Enter a Number'))
if(n//2)*2==n:
    print(f"{n} is Even Number")
else:
    print(f"{n} is Odd Number")
```

```python
# 101                      # 5
# 100                      # 4
# 100

# 1101                     # 13
# 1011                     # 11
# 1001

# using bitwise and (&)
n=int(input('Enter a Number'))
if(n&1)==0:
    print(f"{n} is Even Number")
else:
    print(f"{n} is Odd Number")
```

```python
# using bitwise or (|)
# 101
# 100
# 101

# 1101
# 1011
# 1111

n=int(input('Enter a Number'))
if(n|1)==n+1:
    print(f"{n} is Even Number")
else:
    print(f"{n} is Odd Number")
```

```python
# using shift operator <<  >> 

n=int(input('Enter a Number'))
if(n>>1)<<1==n:
    print(f"{n} is Even Number")
else:
    print(f"{n} is Odd Number")
```
