# factorial

```python
def factorial(n):
    res=1
    for i in range(1,n+1):
        res*=i
    return res

n=int(input("Enter a Number for its factorial"))
print(factorial(n))
```

# Strong Number

### Sum of the factors of the digits of a number is equal to the original number is called Strong Number


```python
def fact(n):
    if n<=0:
        return 1
    else:
        return n * fact(n-1)
    
n=int(input("Enter a Number"))
print("Entered Number is ",n)
sum=0
num=n
while(n>0):
    temp=n%10
    sum=sum+ fact(temp)
    n//=10

print("Strong Number" if sum == num else "Not Strong Number")
```

```python
n=int(input("Enter a Number"))
temp=n
res=0

while n!=0:
    rem=n%10
    fact=1
    for i in range(1,rem+1):
        fact=fact*i
    res = res + fact
    n = n//10

if temp==res:
    print("Strong Number")
else:
    print("Not Strong Number")
```

# Disarium Number

# Automorphic number

```python
n = int(input("a Number "))
sq=n**2
sum=0
d=0
def count(n):
    if n < 10:
        return 1
    else:
        return 1 + count(n // 10)

def rev(n):
    if n < 10:
        return n
    else:
        return (n % 10) * (10 ** (count(n) - 1)) + rev(n // 10)

cnt=count(n)
for i in range(cnt):
    d=sq%10
    sum=sum*10+d
    sq=sq//10
print("Automorphic Number" if rev(sum)== n else "Not Automorphic Number")
```

# Decimal to Hexadecimal

```python
n=int(input("Enter a Decimal Number: "))
x=n
hexa=""

while(n!=0):
    temp=n%16

    if(temp<10):
        hexa=str(temp)+hexa
    else:
        if(temp==10):
            hexa="A"+hexa
        elif(temp==11):
            hexa="B"+hexa
        elif(temp==12):
            hexa="C"+hexa
        elif(temp==13):
            hexa="D"+hexa
        elif(temp==14):
            hexa="E"+hexa
        elif(temp==15):
            hexa="F"+hexa

    n//=16

print(f'The Hexadecimal Value of Decimal Number {x} is {hexa}')
```

# Hexadecimal to Decimal


```python
n=input("Enter a Hexadecimal Number: ").upper()
x=n

dec=0
power=0

for i in range(len(n)-1,-1,-1):

    temp=n[i]

    if(temp>='0' and temp<='9'):
        value=int(temp)
    elif(temp=='A'):
        value=10
    elif(temp=='B'):
        value=11
    elif(temp=='C'):
        value=12
    elif(temp=='D'):
        value=13
    elif(temp=='E'):
        value=14
    elif(temp=='F'):
        value=15

    dec=dec+value*(16**power)
    power+=1

print(f'The Decimal Value of Hexadecimal Number {x} is {dec}')
```

# HCF and LCM

# HCF

```python
n1=int(input("Enter first number"))
n2=int(input("Enter second number"))
hcf=1
for i in range(1,min(n1,n2)+1):
    if n1%i==0 and n2%i==0:
        hcf=i
print(f"HCF of {n1} and {n2} is {hcf}")

print(f'LCM of {n1} and {n2} is {(n1*n2)//hcf}')
```

# For arbitary n Argumnets


```python
n=int(input("How many numbers do you want to enter: "))

nums=[]

for i in range(n):
    num=int(input(f"Enter Number {i+1}: "))
    nums.append(num)

hcf=1

for i in range(1,min(nums)+1):

    flag=True

    for j in nums:
        if j%i!=0:
            flag=False
            break

    if flag:
        hcf=i

print(f"HCF of given numbers is {hcf}")
```

# LCM

```python
n1=int(input("Enter first number"))
n2=int(input("Enter second number"))
lcm=None
loop_counter=0
i=max(n1,n2)
while True:
    loop_counter+=1
    if i%n1==0 and i%n2==0:
        lcm=i
        break
    i+=1
print(f'LCM of {n1} and {n2} is {lcm}')
print(f'Loop Counter = {loop_counter}')
```
